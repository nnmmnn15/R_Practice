### 한 화면에 차트 여러개 그려서 비교하기

```r
par(bg = "white")
price <- runif(10, 1, 100)
plot(price)
```

```r
# type 속성으로 비교하기
# mfrow = 여러 그래프의 수 (2, 2) = 2*2 = 4
par(mfrow = c(2, 2), bg = "white")
plot(price, type= "l")
plot(price, type= "o")
plot(price, type= "h")
barplot(price)
```

### chart 2개를 겹쳐서 출력하기

```r
x <- 1:10

mycol <- rgb(1, 0, 0, seq(0,1,0.1)) # seq를 추가해서 그라데이션 효과처럼
par(bg = "white")
dotchart(x, col = "red")
par(new = T) # 현재 차트를 유지하고 다른 차트를 생성
barplot(
    x,
    col = mycol # red, green, blue, alpha
)
```

### 기타 그래프

```r
x <- c(1,3,6,8,9)
y <- c(12,56,78,32,9)
par(bg = "white", family = "AppleGothic")
plot(x,y)

# 선분 그리기 (3,3)과 (4,4)
segments(6,78,8,32)

# 화살표 그리기
arrows(3, 56, 1, 12, col = "blue")

# 사각형 그리기
rect(4,20,6,30, density = 30, col = "red")

# 글자 쓰기 # srt = 회전
text(4, 40, "This is a 샘플", srt = 50)

# 테두리 활용하기
# side 
#       3  
# 2             4
#       1
mtext("Character", side = 4, adj = 1)

# x축의 추가 : y축의 40위치에 0~10까지 표시하고 싶은 색은 빨간색
axis(1, pos = 40, at = 0:10, col = "red")

# y축의 추가 : y축의 6위치에 20~60까지 표시하고 싶은 색은 파란색
axis(2, pos = 6, at = 20:60, col = "blue")
```
