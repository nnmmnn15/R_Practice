### 선그래프

```r
# 3종류 데이터의 선그래프
abc <- c(260, 300, 250, 280, 310)
def <- c(180, 200, 210, 190, 170)
ghi <- c(210, 250, 260, 210, 270)
```

```r
par(family = "AppleGothic", bg = "white")
plot(
    abc,
    type = "o",
    col = "red",
    ylim = c(0, 400),
    axes = F, # x, y의 좌표 프레임 제거
    ann = F, # x, y의 Label 제거
)

# x 축과 x축의 라벨을 설정
axis(1, at = 1:5, lab = c("서울", "대전", "대구", "광주", "원주"))

# y 축과 y축의 라벨을 설정
axis(2, ylim = c(0, 400))

# main title 지정
title(main = "과일 판매량", col.main = "red", cex.main = 2) # 색, 글자크기

# x축 title 지정
title(xlab = "지역", col.lab = "blue", cex.lab = 2)

# y축 title 지정
title(ylab = "판매량", col.lab = "blue")


# plot 그래프 위에 선을 그림
lines(
    def,
    type = "o",
    col = "green",
    lty = 2
)

lines(
    ghi,
    type = "o",
    col = "blue",
    lty = 2,
    pch = 22,
)

# 범례 legend
legend(
    4, 
    400,
    c("야구장", "축구장", "해변가"),
    col = c("red", "green", "blue"),
    pch = c(21, 21, 22),
    lty = c(1, 2, 2),
    cex = 0.8
)
```

### 학력별 점수 차이

```r
data <- read.csv("../Data/education.csv", header = F)
head(data)
```

```r
# V1 : 남, 여
# V2 : 점수
# V3 : 국어, 영어, 수학
# V4 : 초, 중, 고, 대, 원

data$V1 <- factor(data$V1, levels = 1:2, labels = c("남", "여"))
data$V3 <- factor(data$V3, levels = 1:3, labels = c("국어", "영어", "수학"))
data$V4 <- factor(data$V4, levels = 1:5, labels = c("초", "중", "고", "대", "원"))
```

```r
head(data)
```

```r
str(data)
```

```r
# 컬럼 이름 변경 (성별, 점수, 과목, 학력)
colnames(data) <- c("성별", "점수", "과목", "학력")
head(data)
```

```r
# 점수에 빈도수를 구하여 막대 그래프로 시각화
table(data$`점수`)
max(table(data$`점수`))
```

```r
par(bg = "white", family = "AppleGothic")
barplot(
    table(data$`점수`),
    ylim = c(0, 12),
    xlab = "점수",
    ylab = "빈도수",
    main = "점수별 빈도수",
    cex.main = 2,
)
```

```r
# 성별 학력별 빈도수를 막대그래프로 표현
table(data$`성별`)
table(data$`학력`)
table(data$`성별`, data$`학력`)
table(data$`학력`, data$`성별`)
```

```r
par(bg = "white", family = "AppleGothic")
barplot(
    table(data$`성별`, data$`학력`),
    beside = T,
    xlab = "학력별",
    ylab = "빈도수",
    legend.text = T,
    ylim = c(0, 6),
    col = ifelse(rownames(table(data$`성별`, data$`학력`)) == "남",
        "orange", "yellow"
    )
)
```

```r
rownames(table(data$`성별`, data$`학력`))
rownames(table(data$`성별`, data$`학력`))
```

### 파이차트

```r
sales <- c(210,110,400,550,600,130)
par(bg = "white")
pie(sales)
```

```r
par(bg = "white", family = "AppleGothic")
sales <- c(210,110,400,550,600,130)
week <- c("월", "화", "수", "목", "금", "토")
ratio <- sales / sum(sales) * 100
label <- paste(week, "\n", ratio, "%")

pie(
    sales,
    main = "영업 실적 파이 차트",
    col = rainbow(length(sales)),
    labels = label,
)
```

```r
par(bg = "white", family = "AppleGothic")
sales <- c(210,110,400,550,600,130)
week <- c("월", "화", "수", "목", "금", "토")
ratio <- sales / sum(sales) * 100
label <- paste(week, "\n", ratio, "%")

pie3D(
    sales,
    main = "영업 실적 파이 차트",
    col = rainbow(length(sales)),
    labels = label,
    explode = 0.05,
    labelcex = 1.8
)
```
