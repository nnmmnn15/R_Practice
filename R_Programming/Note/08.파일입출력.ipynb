### 세계 GDP순위 데이터 정제 및 차트 만들기

```r
# 데이터 불러오기
GDP_ranking <- read.csv("../Data/GDP.csv")
GDP_ranking
head(GDP_ranking)
```

```r
# GDP_ranking2 <- GDP_ranking[-(1:4), -c(3, 6)]
GDP_ranking2 <- GDP_ranking[-(1:4), c(1, 2, 4, 5)]
head(GDP_ranking2)
```

```r
# 상위 15개국만 선별 한다.
GDP_ranking15 <- head(GDP_ranking2, 15)

GDP_ranking15
```

```r
# 컬럼 이름 변경
colnames(GDP_ranking15) <- c("Code", "Ranking", "Nation", "GDP")
GDP_ranking15
```

```r
# rownames 새로 구성하기
rownames(GDP_ranking15) <- c()
GDP_ranking15
print(GDP_ranking15)
```

```r
library(stringr)
```

```r
# GDP 칼럼을 숫자로 변경하기
GDP_ranking15$GDP <- as.numeric(str_remove_all(GDP_ranking15$GDP, ","))
GDP_ranking15
```

```r
# 지수 없애기
options(scipen = 100)
# 그래프 환경설정
options(warn =-1)
options(repr.plot.width=15, repr.plot.height=9)
```

```r
# 막대 그래프의 시각화
# 차트의 한글 폰트
par(family = "AppleGothic", bg = "white")
barplot(
    GDP_ranking15$GDP, # GDP상위 15개국 Data
    col = rainbow(nrow(GDP_ranking15)), # color
    # col = c("red", "blue", "green", "yellow", "purple", "gray", "pink")
    xlab = "국가(National)", # x축 이름
    ylab = "단위(달러)", # y축 이름
    main = "2018년 GDP 세계 15위 국가", # Title
    names = GDP_ranking15$Nation,
    cex.names = 0.65,
)
```

```r
# 막대 그래프의 시각화
# 차트의 한글 폰트
par(family = "AppleGothic", bg = "white")
barplot(
    GDP_ranking15$GDP, # GDP상위 15개국 Data
    col = rainbow(nrow(GDP_ranking15)), # color
    # col = c("red", "blue", "green", "yellow", "purple", "gray", "pink")
    xlab = "국가(National)", # x축 이름
    ylab = "단위(달러)", # y축 이름
    main = "2018년 GDP 세계 15위 국가", # Title
    names = GDP_ranking15$Code,
)
```

```r
colorDefind <- function() {
    return(ifelse(
        ifelse(GDP_ranking15$Code == c("USA"), 1, 0) +
            ifelse(GDP_ranking15$Code == c("CHN"), 1, 0) +
            ifelse(GDP_ranking15$Code == c("KOR"), 1, 0) == 1, "red", "blue"
    ))
}

# ifelse(
#     ifelse(GDP_ranking15$Code == c("USA"), 1, 0) +
#         ifelse(GDP_ranking15$Code == c("CHN"), 1, 0) +
#         ifelse(GDP_ranking15$Code == c("KOR"), 1, 0) == 1, "red", "blue"
# )
```

```r
# 미국과 중국 한국만 빨간색으로 나머지는 파란색으로 표현하기
# 막대 그래프의 시각화
# 차트의 한글 폰트
par(family = "AppleGothic", bg = "white")
barplot(
    GDP_ranking15$GDP, # GDP상위 15개국 Data
    # col = c(ifelse(GDP_ranking15$Code == "USA" | GDP_ranking15$Code == "KOR" |
    #     GDP_ranking15$Code == "CHN",
    #     "red", "blue"
    # )), # color
    col = colorDefind(),
    xlab = "국가(National)", # x축 이름
    ylab = "단위(달러)", # y축 이름
    main = "2018년 GDP 세계 15위 국가", # Title
    names = GDP_ranking15$Code,
)
```

```r
# 주의 반복 비교 이기 때문에 값의 순서가 달라지면 완전 다른결과가 나올수 있음
c(ifelse(GDP_ranking15$Code == c("USA", "CHN", "KOR"), 1, 0))
```

```r
# 단위를 천단위로 표시
# 미국과 중국 한국만 빨간색으로 나머지는 파란색으로 표현하기
# 막대 그래프의 시각화
# 차트의 한글 폰트
par(family = "AppleGothic", bg = "white")
barplot(
    GDP_ranking15$GDP/1000, # GDP상위 15개국 Data 나누기 1000
    # col = c(ifelse(GDP_ranking15$Code == "USA" | GDP_ranking15$Code == "KOR" |
    #     GDP_ranking15$Code == "CHN",
    #     "red", "blue"
    # )), # color
    col = colorDefind(),
    xlab = "국가(National)", # x축 이름
    ylab = "단위(천달러, 1000$)", # y축 이름
    main = "2018년 GDP 세계 15위 국가", # Title
    names = GDP_ranking15$Code,
)
```

```r
# GDP가 3000이상인 국가만 빨간색으로 표시하기
par(family = "AppleGothic", bg = "white")
barplot(
    GDP_ranking15$GDP/1000, # GDP상위 15개국 Data 나누기 1000
    # col = c(ifelse(GDP_ranking15$Code == "USA" | GDP_ranking15$Code == "KOR" |
    #     GDP_ranking15$Code == "CHN",
    #     "red", "blue"
    # )), # color
    col = ifelse(GDP_ranking15$GDP/1000 >= 3000, "red", "blue"),
    xlab = "국가(National)", # x축 이름
    ylab = "단위(천달러, 1000$)", # y축 이름
    main = "2018년 GDP 세계 15위 국가", # Title
    names = GDP_ranking15$Code,
    ylim = c(0, 25000)
)
```

### Sampling

```r
set.seed(42) # 랜덤 시드 고정
sample(1:100, 10)
nrow(iris)
```

```r
set.seed(42) # 랜덤 시드
samp <- sample(1:nrow(iris), nrow(iris) * 0.7)
samp
```

```r
irisSamp <- iris[samp,]
head(irisSamp)
```

```r
# Sampling 검증 : Species빈도수 // 데이터 크기 검증
table(irisSamp$Species)
```

```r
samp <- c(sample(1:50, 35), sample(51:100, 35), sample(101:150, 35))
```

```r
irisSamp <- iris[samp, ]
head(irisSamp)
# Sampling 검증 : Species빈도수 // 데이터 크기 검증
table(irisSamp$Species)
```

```r
# train Data
iris.train <- iris[samp,]
head(iris.train)
```

```r
# test Data
iris.test <- iris[-samp,]
head(iris.test)
```

```r
# 데이터 크기 검증
table(iris.test$Species)
```

```r
# 데이터 검증
# train과 test의 median 과 mean 값의 유사도 확인
summary(iris.train)
summary(iris.test)
```
