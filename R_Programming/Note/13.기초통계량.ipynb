### 수학 함수의 시각화

: y = x^2 // y는 x의 제곱

```r
# 
f <- function(x){
    return(x^2)
}
```

```r
x <- 1:5
f(x)
```

```r
x <- seq(-5, 5, 0.1)

par(bg = "white")

plot(
    x,
    f(x),
    type = "l"
)
```

```r
# y = x^2 + 2x + 3
f <- function(x){
    return(x^2 + 2*x + 3)
}
```

```r
x <- seq(-5, 5, 0.1)
```

```r
par(bg = "white")
plot(
    x,
    f(x),
    type = "l",
    xlim = c(-6, 6)
)
```

```r
# y = e^-x + 4x^2 + x + 3|x| - 3
# exp(-x)
f <- function(x){
    return (exp(-x) + 4*(x^2) + x + 3 + abs(x) - 3)
}
```

```r
par(bg = "white")
plot(
    x,
    f(x),
    type = "l",
)
```

### 기초 통계량

```r
# 커피
cafe <- read.csv("../Data/coffee.csv")
head(cafe)
tail(cafe)
```

```r
# 커피의 빈도수
table(cafe$Coffee)
```

```r
# 범위 빈도수
coffee_range <- seq(0, 60, 10) # 범위 지정
coffee_range
table(cut(cafe$Coffee, breaks = coffee_range, right=F)) # ( open , ] close
```

```r
# 줄기-잎 그림
stem(cafe$Coffee)
# 0 | 35589 => 3, 5, 5, 8, 9
# 1 | 011348 => 10, 11, 11, 13, 14, 18
```

```r
# 커피와 쥬스판매량의 변동계수를 구하여 메뉴를 선택하자
cat("커피의 변동계수는 ",sd(cafe$Coffee) / mean(cafe$Coffee),"\n")
cat("쥬스의 변동계수는 ",sd(cafe$Fruit.Cup) / mean(cafe$Fruit.Cup),"이므로 \n")

ifelse(sd(cafe$Coffee) / mean(cafe$Coffee) > sd(cafe$Fruit.Cup) / mean(cafe$Fruit.Cup), paste("쥬스"), paste("커피"))
```
